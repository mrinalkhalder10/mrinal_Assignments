**Scipy:**

**We have the min and max temperatures in a city In India for each months of the year.**

**We would like to find a function to describe this and show it graphically, the datase given below.**

**Task:**

**1.fitting it to the periodic function**

**2.plot the fit Data**

**Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25**

**Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18**

```
import numpy as np

tempt_max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
tempt_min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

import matplotlib.pyplot as plt
%matplotlib inline
months = np.arange(12)

from scipy import optimize
def yearly_temperature(times, avg, ampl, time_offset):
    return (avg
            + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temperature, months,
                                      tempt_max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temperature, months,
                                      tempt_min, [-40, 20, 0])


days = np.linspace(0, 12, num=365)
plt.figure()
plt.plot(months, tempt_max, 'ro')
plt.plot(days, yearly_temperature(days, *res_max), 'r-')
plt.plot(months, tempt_min, 'bo')
plt.plot(days, yearly_temperature(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

**Charts to plot:**

**2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
df =pd.read_csv(data)
d=df['sex'].value_counts()
d.plot.pie(subplots = True,figsize = (7, 7), fontsize = 20, autopct = '%.2f%%',shadow=True,labels=['Male','Female'],label='Sex')
plt.show()
```

```
category1 = df[df.sex=='male'].plot.scatter( 'fare','age',color='blue',label='male') 
df[df.sex=='female'].plot.scatter('fare','age', color='red',label='female',alpha=0.5,ax=category1) 
plt.xlabel('Fare Paid')
plt.ylabel('Age')
plt.show()
```
