```python
df=pd.read_csv('C:/Users/hp/Music/iNeuron/titanic data/train.csv')

df
```

```python
df1 = df[['Survived','Name','Pclass','Sex','Age','SibSp','Parch','Fare']]
```

```python
df1['Surname'] = df1.Name.str.split(',').str.get(0)
df1
```

```python
grp = df1.groupby(['Sex', 'Pclass'])
df1.Age = grp.Age.apply(lambda x: x.fillna(x.median()))
df1
```

```python
#df1[['Age']]=df1[['Age']].fillna(value=age_mean)
df1['Age'] = df1['Age'].astype(int)
```

```python
df1.isnull().sum()
```

```python
#Categorical Sex column to numeric

from sklearn.preprocessing import LabelEncoder

number = LabelEncoder()
df1['Sex'] = number.fit_transform(df1['Sex'].astype('str'))
df1
# 1-male : 0-female
```

```python
print(df1.Pclass.unique())
print(df1.Sex.unique())
print(df1.Age.unique())
print(df1.SibSp.unique())
print(df1.Parch.unique())
```

```python
y=df1[['Survived']]
X=df1.drop(df1[['Survived','Surname','Name']],axis=1)
```

```python
print(X)
print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.25, random_state= 355)
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
scalar = StandardScaler()

x_transform = scalar.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_transform,y,test_size = 0.25, random_state= 355)
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
clf.score(x_test,y_test)
```

```python
grid_param = {
    'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,32,1),
    'min_samples_leaf' : range(1,10,1),
    'min_samples_split': range(2,10,1),
    'splitter' : ['best', 'random']
    
}
```

```python
grid_search = GridSearchCV(estimator=clf,
                     param_grid=grid_param,
                     cv=5,
                    n_jobs =-1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_parameters = grid_search.best_params_
print(best_parameters)
```

```python
grid_search.best_score_
```

```python
clf = DecisionTreeClassifier(criterion = 'gini', max_depth =10, min_samples_leaf= 3, min_samples_split= 3, splitter ='random')
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```
